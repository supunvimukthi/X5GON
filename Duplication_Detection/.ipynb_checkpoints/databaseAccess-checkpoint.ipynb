```python
""" word distribution of urls """
import re
domain_words_array=[]
for domain in url_domains:
    domain_words=dict()
    for url in url_domains[domain]:
        for word in url.split("/"):
            if word.isalnum():
                if word not in domain_words and word !="" and not word.isdigit() and word not in ["-","%","_","=","?"]:
                    domain_words[word]=0
                if word !="" and not word.isdigit() and word not in ["-","%","_","=","?"]:
                    domain_words[word]+=1
            prev=""
            for x in re.split(r'(%|/?/|-|_|=)', word):
#                 if x not in domain_words and x !="" and not x.isdigit() and x not in ["-","%","_","=","?"]:
#                     domain_words[x]=0
#                 if x !="" and not x.isdigit() and x not in ["-","%","_","=","?"]:
#                     domain_words[x]+=1
                if x.isalnum():
                    bigram=prev+" "+x
                    if bigram!="":
                        if bigram not in domain_words and not bigram.strip().isdigit():
                            domain_words[bigram]=0
                        if not bigram.strip().isdigit():
                            domain_words[bigram]+=1
                    prev=x
    domain_words_array.append(domain_words)
```

```python
""" match meta data with the data from the database --> title and author """

final=[]
count=0
queries=dict()
for i in meta_data_table:
    if i[2] not in queries:
        queries[i[2]]=[i[0],i[1]]

for j in x:
    if j[0] in queries:
        try:
            
            queries[j[0]].append(j[1]["Title"])
            queries[j[0]].append(j[1]["Author"])
            count+=1
        except:
            try:
                queries[j[0]].append("")
                queries[j[0]].append(j[1]["Author"])
            except:
                pass
                
final=[[key]+queries[key] for key in queries]
len(final),count
```
